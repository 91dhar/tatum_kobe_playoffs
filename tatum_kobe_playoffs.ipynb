```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df.sort_values(by='Career Game', inplace=True)

# Creating a list of sections
sections = [(i, i + 9) for i in range(1, 95, 10)]  # Sections: [(1, 10), (11, 20), ... , (81, 90), (91, 94)]

# Initializing a dictionary to store the cumulative total points
cumulative_totals = {}

# Initializing a variable to keep track of cumulative total
cumulative_total = 0

# Iterating through the sections and calculate cumulative total points
for section_start, section_end in sections:
    section_name = f"{section_start}-{section_end}"
    section_df = df[(df['Career Game'] >= section_start) & (df['Career Game'] <= section_end)]
    total_points = section_df['PTS'].sum()
    cumulative_total += total_points
    cumulative_totals[section_name] = cumulative_total
print(cumulative_df)

# Creating a DataFrame from the cumulative_totals dictionary
cumulative_df = pd.DataFrame.from_dict(cumulative_totals, orient='index', columns=['Cumulative Total Points'])

# Calculating the trendline using numpy.polyfit
x_values = np.arange(len(cumulative_df))
y_values = cumulative_df['Cumulative Total Points']
trendline = np.polyfit(x_values, y_values, 1)  # Linear trendline

# Creating a bar graph
plt.figure(figsize=(12, 6))
plt.bar(cumulative_df.index, cumulative_df['Cumulative Total Points'], color='skyblue', label='Cumulative Total Points')
plt.plot(x_values, trendline[0] * x_values + trendline[1], color='orange', linestyle='--', label='Trendline')
plt.xlabel('Career Game Sections')
plt.ylabel('Cumulative Total Points')
plt.title('Cumulative Total Points in Career Game Sections for Jayson Tatum with Trendline')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Creating the plot for turnovers
plt.figure(figsize=(12, 6))
plt.plot(df['Career Game'], df['TOV'], marker='o', linestyle='-', color='g', label='Turnovers')
plt.xlabel('Career Game')
plt.ylabel('Turnovers')
plt.title('Jayson Tatum - Turnovers - Playoffs Career')

# Calculating the degree of the polynomial trendline (you can adjust this)
degree = 5

# Defining the number of games in each trend section
games_per_section = 10

# Fitting the polynomial trendline
x_values = df['Career Game']
y_values = df['TOV']
coefficients = np.polyfit(x_values, y_values, degree)
polynomial = np.poly1d(coefficients)

# Creating x values for the trendline (with 10-game increments)
x_trendline = np.arange(x_values.min(), x_values.max() + 1, games_per_section)

# Calculating the y values for the trendline
y_trendline = polynomial(x_trendline)

# Adding the trendline to the plot
plt.plot(x_trendline, y_trendline, color='orange', linestyle='--', label=f'Polynomial Trendline (Degree {degree})')
x_tick_values = np.arange(x_values.min(), x_values.max() + 1, 10)
plt.xticks(x_tick_values)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

tatum_df = pd.read_excel('tatum_playoffs.xlsx')

kobe_df = pd.read_excel('kobe_playoffs.xlsx')

# Create a figure for the plot
plt.figure(figsize=(12, 6))

# Define the number of bars and bar width
num_bars = len(tatum_df)
bar_width = 0.35

# Create an array of x values for the bars
x_values = np.arange(num_bars)

# Plot the grouped barplots for career points for both players
plt.bar(x_values - bar_width/2, tatum_df['PTS'], bar_width, label='Jayson Tatum', color='green')
plt.bar(x_values + bar_width/2, kobe_df['PTS'], bar_width, label='Kobe Bryant', color='blue')

# Calculate the trendline for Jayson Tatum's data
degree_tatum = 10
coefficients_tatum = np.polyfit(x_values, tatum_df['PTS'], degree_tatum)
polynomial_tatum = np.poly1d(coefficients_tatum)
y_trendline_tatum = polynomial_tatum(x_values)

# Calculate the trendline for Kobe Bryant's data
degree_kobe = 10
coefficients_kobe = np.polyfit(x_values, kobe_df['PTS'], degree_kobe)
polynomial_kobe = np.poly1d(coefficients_kobe)
y_trendline_kobe = polynomial_kobe(x_values)

# Plot the trendlines
plt.plot(x_values, y_trendline_tatum, color='black', linestyle='--', label=f'Tatum Trendline (Degree {degree_tatum})')
plt.plot(x_values, y_trendline_kobe, color='purple', linestyle='--', label=f'Kobe Trendline (Degree {degree_kobe})')

# Add labels, title, and legend
plt.xlabel('Career Game')
plt.ylabel('Points')
plt.title('Playoffs Career Points - Grouped Barplots with Trendlines')
plt.grid(True)
plt.legend()

# Set x-axis tick values in a sequence of 10 games
x_tick_values = np.arange(0, num_bars, 10)  # Start at 0, increment by 10
x_tick_labels = x_tick_values + 1

plt.xticks(x_tick_values, x_tick_labels)

plt.show()
```

```python
import pandas as pd

tatum_df = pd.read_excel('tatum_playoffs.xlsx')

kobe_df = pd.read_excel('kobe_playoffs.xlsx')

# Calculate cumulative sum of points for each player
tatum_df['Cumulative Points'] = tatum_df['PTS'].cumsum()
kobe_df['Cumulative Points'] = kobe_df['PTS'].cumsum()

# Calculate cumulative points after every 10 games
tatum_cumulative_10_games = tatum_df.groupby(tatum_df.index // 10 * 10)['Cumulative Points'].last()
kobe_cumulative_10_games = kobe_df.groupby(kobe_df.index // 10 * 10)['Cumulative Points'].last()

# Create a DataFrame to store the cumulative points comparison after every 10 games
cumulative_points_10_games_df = pd.DataFrame({
    'Games': tatum_cumulative_10_games.index + 1,
    'Tatum Cumulative Points': tatum_cumulative_10_games.values,
    'Kobe Cumulative Points': kobe_cumulative_10_games.values
})

print(cumulative_points_10_games_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Creating a figure and axis for the chart
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the cumulative points data as a stem chart with stacked bars and values on top
stem_tatum = ax.stem(cumulative_points_10_games_df['Games'], cumulative_points_10_games_df['Tatum Cumulative Points'], linefmt='g-', markerfmt='go', basefmt=' ')
stem_kobe = ax.stem(cumulative_points_10_games_df['Games'], cumulative_points_10_games_df['Kobe Cumulative Points'], linefmt='b-', markerfmt='bo', basefmt=' ')

# Add labels and title to the chart
ax.set_xlabel('Games (Every 10 Games)')
ax.set_ylabel('Cumulative Points')
ax.set_title('Cumulative Points Comparison After Every 10 Games')

# Add values on top of each stem for Tatum
for i, val in enumerate(cumulative_points_10_games_df['Tatum Cumulative Points']):
    ax.annotate(str(val), xy=(cumulative_points_10_games_df['Games'][i], val), xytext=(0, 5), textcoords='offset points', va='bottom', ha='center')

# Add values on top of each stem for Kobe
for i, val in enumerate(cumulative_points_10_games_df['Kobe Cumulative Points']):
    ax.annotate(str(val), xy=(cumulative_points_10_games_df['Games'][i], val), xytext=(0, 5), textcoords='offset points', va='bottom', ha='center')

# Fit a polynomial trendline (you can adjust the degree)
degree = 2
x_values = cumulative_points_10_games_df['Games']
y_values_tatum = cumulative_points_10_games_df['Tatum Cumulative Points']
y_values_kobe = cumulative_points_10_games_df['Kobe Cumulative Points']

coefficients_tatum = np.polyfit(x_values, y_values_tatum, degree)
coefficients_kobe = np.polyfit(x_values, y_values_kobe, degree)

polynomial_tatum = np.poly1d(coefficients_tatum)
polynomial_kobe = np.poly1d(coefficients_kobe)

# Create x values for the trendline
x_trendline = np.linspace(x_values.min(), x_values.max(), 100)

# Calculate the y values for the trendline for both players
y_trendline_tatum = polynomial_tatum(x_trendline)
y_trendline_kobe = polynomial_kobe(x_trendline)

# Plot the trendline for Tatum
ax.plot(x_trendline, y_trendline_tatum, color='green', linestyle='--', label=f'Tatum Trendline (Degree {degree})')

# Plot the trendline for Kobe
ax.plot(x_trendline, y_trendline_kobe, color='purple', linestyle='--', label=f'Kobe Trendline (Degree {degree})')

plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Sort the dataframes by 'Career Game'
tatum_df.sort_values(by='Career Game', inplace=True)
kobe_df.sort_values(by='Career Game', inplace=True)

# Creating a list of sections
sections = [(i, i + 9) for i in range(1, 95, 10)]  # Sections: [(1, 10), (11, 20), ... , (81, 90), (91, 94)]

# Initializng dictionaries to store the cumulative total points for each player
cumulative_totals_tatum = {}
cumulative_totals_kobe = {}

# Initializing variables to keep track of cumulative totals
cumulative_total_tatum = 0
cumulative_total_kobe = 0

# Iterating through the sections and calculate cumulative total points for each player
for section_start, section_end in sections:
    section_name = f"{section_start}-{section_end}"
    
    # Filter data for Jayson Tatum
    section_tatum_df = tatum_df[(tatum_df['Career Game'] >= section_start) & (tatum_df['Career Game'] <= section_end)]
    total_points_tatum = section_tatum_df['PTS'].sum()
    cumulative_total_tatum += total_points_tatum
    cumulative_totals_tatum[section_name] = cumulative_total_tatum
    
    # Filter data for Kobe Bryant
    section_kobe_df = kobe_df[(kobe_df['Career Game'] >= section_start) & (kobe_df['Career Game'] <= section_end)]
    total_points_kobe = section_kobe_df['PTS'].sum()
    cumulative_total_kobe += total_points_kobe
    cumulative_totals_kobe[section_name] = cumulative_total_kobe

# Creating DataFrames from the cumulative_totals dictionaries for each player
cumulative_tatum_df = pd.DataFrame.from_dict(cumulative_totals_tatum, orient='index', columns=['Tatum Cumulative Total Points'])
cumulative_kobe_df = pd.DataFrame.from_dict(cumulative_totals_kobe, orient='index', columns=['Kobe Cumulative Total Points'])

# Combine both DataFrames into one
cumulative_df = pd.concat([cumulative_tatum_df, cumulative_kobe_df], axis=1)

# Creating a bar graph for both players
plt.figure(figsize=(10, 6))
bar_width = 0.4
index = range(len(cumulative_df))

plt.bar(index, cumulative_df['Tatum Cumulative Total Points'], width=bar_width, color='green', label='Jayson Tatum')
plt.bar([i + bar_width for i in index], cumulative_df['Kobe Cumulative Total Points'], width=bar_width, color='black', label='Kobe Bryant')

plt.xlabel('Career Game Sections')
plt.ylabel('Cumulative Total Points')
plt.title('Cumulative Total Points in Career Game Sections')
plt.xticks([i + bar_width/2 for i in index], cumulative_df.index, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

# Adding trendlines
x_values = np.array(index)
x_values_combined = np.concatenate((x_values, x_values + bar_width))
y_values_tatum = cumulative_df['Tatum Cumulative Total Points'].values
y_values_kobe = cumulative_df['Kobe Cumulative Total Points'].values

model_tatum = LinearRegression().fit(x_values.reshape(-1, 1), y_values_tatum)
model_kobe = LinearRegression().fit(x_values.reshape(-1, 1), y_values_kobe)

plt.plot(x_values_combined, model_tatum.predict(x_values_combined.reshape(-1, 1)), linestyle='--', color='green', label='Tatum Trendline')
plt.plot(x_values_combined, model_kobe.predict(x_values_combined.reshape(-1, 1)), linestyle='--', color='purple', label='Kobe Trendline')

# Adding values on top of each bar
for i, val in enumerate(cumulative_df['Tatum Cumulative Total Points']):
    plt.text(i, val + 50, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold', color='red')

for i, val in enumerate(cumulative_df['Kobe Cumulative Total Points']):
    plt.text(i + bar_width, val + 50, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold', color='red')

plt.tight_layout()
plt.show()

print(cumulative_df)
```
